# Regex 

Comandos

 - search: encontra as posicoes padroes dentro de uma sting, se estiverem presentes

 - match: encontra se o comeco de uma string e igual a um determinado padrao

 - findall: encontra todas as substrings em uma string que correspondam a um padrao

# Meta caracteres

 - . qualquer caracter (exceto linha nova)
 - \w qualquer caracter alfanumerico
 - \W contrario de \w
 - \d digito
 - \D contrario
 - \s espaco em branco
 - \S contrario
 - ^ comeca com
 - $ termina com
 - \ usado antes para especificar
 
# QUANTIFICADORES

 - []  OPCIONAL ENTRE OS QUE ESTAO DENTRO DO COLCHETES
 - ()  PEGA GRUPO DE CARACTERES
 - '*'  0 A MAIS VEZES
 - ?  0 OU 1
 - '+'  1 OU MAIS
 - {M,N} M A N VEZES
 - |  OU
 
interroga;'ao junto com o + quebra em duas partes.
 
# Exemplos



```python
import re
```

```python
texto1 = 'ola meu numero e (81)57878-7878'
texto2 = 'a minha placa e FTR-1522'
texto3 = 'entre em contato, meu email e teste.teste@teste.com'
texto4 = 'FTR-6469 e minha placa'
texto5 = 'meu numero e (42)7878-2556 e da minha mae e (84)78789-4512'

# search
print('-'*10 + 'search' + '-'*10 )
print( re.search(  '\(\d{2}\)\d{4,5}-\d{4}', texto1).group() )
print( re.search( '[A-Z]{3}-\d{4}', texto2 ).group())
print( re.search( '\S+@\w+\.\w+', texto3).group())
print( '\n' )

# print( re.search( '\S+$@', texto3).group())

# match
print('-'*10 + 'match' + '-'*10 )
print( re.match( '[A-Z]{3}-\d{4}', texto2) )
print( re.match( '[A-Z]{3}-\d{4}', texto4).group() )
print( '\n' )

#findall
print('-'*10 + 'findall' + '-'*10 )
print( re.findall( '\(\d{2}\)\d{4,5}-\d{4}', texto5 ))
print( '\n' )
```

```python
# EXEMPLO 
texto = 'minha casa fica na rua carneiro, 78. o cep e 11111-999 e meu site e https://www.marcosvinicius.com.br'

# url
re.findall( 'https://[A-Za-z0-9./]+', texto )
```

# CDS 

```python
texto1 = 'Rua Armando shibaata, 97 - Vila leopoldina, sao paulo'
texto2 =  'Ab. dos trabalhadores, 403 - centro, mogi guacu'
texto3 = 'Al. Moraes, 403 - vila golmes, Pernambuco'

text = [texto1, texto2, texto3]

for t in text:
    # nome da rua
    print( re.match( '(.+?),', t ).group() )

    # numero de residencia
    print( re.search( '\d+', t).group())
    
    # bairro
    # obs quando bota () ele so pega oque esta dentro coloca group(1) o 0 pega com virgula e -
    print( re.search( '-(.+),', t).group(1))
    
    # nome do estado
    print( re.search( ',(.[a-zA-Z].+)', t).group(1), '\n\n')
```

```python
texto1 = 'casa o mundo esta mudando. Tecnologias utilizadas na atualizadade estao se tornando as tarefas cada vez mais mecanizadas e preticas, onde a mente humana e deixada livre para exercer o papel criativo relual contrivial centrix chusc'

# extrair todas as palavras que possuem de  5 a 9 letras 
print( re.findall( '\w{5,9}', texto1 ), '\n' )

# extrair todos os trechos separados por ponto
print( re.findall( '.+?\.', texto1), '\n')

#extrair todas as palavras que contenham nd e todas palvras que contenham a ou i
print( re.findall( '\S+nd\S+', texto1), '\n' )
print( re.findall( '\S+[ai]\S+', texto1), '\n' )

# extrair todas as palavras que tenham c
print( re.findall( '(^c\S+|\S+c\S+|\S+c$|c\S+)', texto1 ))
```
